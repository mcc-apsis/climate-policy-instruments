```python
adjective_list = []
adjective_dict = {}
from bs4 import BeautifulSoup
import pycountry
import re

# read in the table with adjectives for each country
with open("../data/country_adjectives.html","r") as f:
    soup = BeautifulSoup(f)

from geotext import GeoText

# define missing countries to be translated
missing_countries = {
    "Burma": "Myanmar",
    "Democratic Republic of the Congo": "Congo, The Democratic Republic of the",
    "East Timor": "Timor-Leste",
    "England": "United Kingdom",
    "European Union": None,
    "Ivory Coast": "CÃ´te d'Ivoire",
    "Laos": "Lao",
    "Macau": "Macao",
    "Pitcairn Islands": "Pitcairn",
    "Sahrawi": "Western Sahara",
    "Scotland": "United Kingdom",
    "Somaliland": "Somalia",
    "South Ossetia": None,
    "United States Virgin Islands": "Virgin Islands, U.S."
}
    
# Parse the table
table = soup.find("table")
for row in table.findAll("tr"):
    cols = row.findAll("td")
    if len(cols) > 0:
        cname = cols[0].find("a").getText()
        try:
            c = pycountry.countries.search_fuzzy(cname)[0]
        except:
            c_alt = missing_countries[cname]
            if c_alt is not None:
                c = pycountry.countries.search_fuzzy(c_alt)[0]
            else:
                continue
        if len(cols[1].findAll("li"))>0:
            adjectives = [x.getText() for x in cols[1].findAll("li")]
        else:
            adjectives = [cols[1].getText()]
        adjectives = [re.sub("\[.*\]","",x) for x in adjectives]
        for a in adjectives:
            adjective_list.append(a)
            adjective_dict[a] = c
        
        
        
    
{k: adjective_dict[k] for k in list(adjective_dict)[:5]}
```

```python
places = []

import re

# Go through the rows of the dataframe
for i, row in df.iterrows():
    
    # Get the text we want to geoparse, join title and abstract, get rid of copyright stuff
    t = str(row['title']) + " " + str(row['abstract'])
    t = t.split("Copyright (C)")[0] 
    t = re.split("\([C-c]\) [1-2][0-9]{3} Elsevier",t)[0] 
    t = t.split("Published by Elsevier")[0] 
    t = t.split("Copyright. (C)")[0] 
    t = re.split("\. \(C\) [1-2][0-9]{3} ",t)[0] 
    t = re.split("\. \(C\) Copyright",t)[0] 
    
    t = re.sub('\\([^()\\d]*\\d[^()]*\\)', '',t) # all citations in parentheses
    t = re.sub('ACKNOWLEDMENTS.*', '',t)
    t = re.sub('Acknowledgements.*', '',t)
    t = re.sub('REFERENCES.*', '',t)
    t = re.sub('Correspondence.*', '',t)
    
    for adj in adjective_list:
        if adj in t:
            places.append({
                "doc_id":row["id"],"country_code3": adjective_dict[adj].alpha_3,
                "feature_class": "A", "feature_code": "PCLI", "word": adj,
                "country_conf": 0.95
            })
            
place_df = pd.DataFrame.from_dict(places)
print(place_df.shape)
print(place_df.doc_id.unique().shape)

place_df.to_csv('../data/place_adjectives.csv',index=False)

place_df.head()


    
    
```

```python
true_countries = (pd.read_csv('../data/doc_countries.csv')[['doc_id','countries__code3']]
      .drop_duplicates()
      .rename(columns={'doc_id':'nacsos_id'})
     )
print(true_countries.shape)
true_countries = pd.merge(true_countries, labels[['id','nacsos_id','INCLUDE']])
true_countries = true_countries[true_countries['INCLUDE']>=0.5]
true_countries = true_countries[true_countries['id'].str.contains('W')]
true_countries = true_countries.rename(columns={
    'countries__code3': 'true_country',
    'nacsos_id': 'doc_id'
})[['doc_id', 'true_country']]
true_countries.head()
```

```python
pred_countries = places[['doc_id','country_code3']].drop_duplicates()
rel_labels = labels[labels['INCLUDE']>0.5]
pred_countries = pred_countries[pred_countries['doc_id'].isin(rel_labels['nacsos_id'])]
pred_countries = pred_countries.rename(columns={'country_code3': 'pred_country'})
pred_countries
```

```python
ac = pd.merge(
    true_countries, pred_countries, 
    left_on=['doc_id','true_country'],
    right_on=['doc_id','pred_country'],
    how='outer'
)
ac
```

```python
def prec_recall(ac):
    tps = ac[
        (pd.notna(ac['true_country'])) &
        (pd.notna(ac['pred_country']))
    ].shape[0]
    fps = ac[
        (pd.isna(ac['true_country'])) &
        (pd.notna(ac['pred_country']))
    ].shape[0]
    fns = ac[
        (pd.notna(ac['true_country'])) &
        (pd.isna(ac['pred_country']))
    ].shape[0]
    p = tps / (tps+fps)
    r = tps / (tps+fns)
    return p, r

p, r = prec_recall(ac)
print(f'precision: {p}, recall: {r}')
```
