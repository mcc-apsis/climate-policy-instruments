```python
import pandas as pd
import re
import numpy as np

df = (pd.read_feather('../data/0_labelled_documents.feather')
               .sort_values('id')
               .sample(frac=1, random_state=1)
               .reset_index(drop=True)
              )
df["seen"] = 1
df.head()
```

```python
y_prefix = "4 -"
min_samples = 10

targets = [x for x in df.columns if re.match(f"^{y_prefix}",x)]
targets = [x for x in targets if df[x].sum()>min_samples]
df = df.replace(-1,np.NaN)
df = df.replace(2, 1)
df = df.dropna(subset=targets).reset_index(drop=True)
if "content" in df.columns:
    tcol = "content"
else:
    tcol = "abstract"
df = df.dropna(subset=["title", tcol]).reset_index(drop=True)

if y_prefix=="INCLUDE":
    nonrandom_index = df[df["representative_sample"]==0].index
    df['labels'] = df[targets[0]]
else:
    df = df[df["INCLUDE"]==1].reset_index(drop=True)
    nonrandom_index = df[df["representative_relevant"]==0].index
    df['labels'] = list(df[targets].values.astype(int))

print(targets)
df
```
